```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
#Example 3.4. pag. 84 
def h(x):
    return (np.cos(50*x)+np.sin(20*x))**2
x=np.linspace(0,1,100)
plt.plot(x,h(x))
plt.show()
n=10**4
U=np.random.uniform(0,1,size=n)
plt.hist(h(U))
plt.show()
I1=[]
I0=[]
I2=[]
for i in range(100,len(U)):
    I1.append(np.mean(h(U[:i]))+np.std(h(U[:i])))
    I2.append(np.mean(h(U[:i]))-np.std(h(U[:i])))
    I0.append(np.mean(h(U[:i])))
plt.plot(I0)
plt.plot(I1,'--')
plt.plot(I2,'--')
plt.show()
```
