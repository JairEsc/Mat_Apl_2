```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from pandas.plotting import table
```

```python
sheet_names=['QUERETARO','CANCUN','MONTERREY']
```

```python
metrics=['CA','PLAND','LPI','ED','PAFRAC','TCA','CPLAND']
columns=['1985-1995','1986-2005','2006-2015']
limits_y=[[0,4,0.5],[0,45000,5000],[0,9,1]]
```

```python
city='QUERETARO'
A=pd.read_excel('grafs_luz_2.xlsx',sheet_name=city)
for indexes_i in range(3):
  indexes=[[1,2],[0,5],[3,4,6]][indexes_i]
  metrics_i=[metrics[r] for r in indexes]
  limits_y_=limits_y[indexes_i]
  B=A.loc[indexes, '1985-1995':'2006-2015']
  B.index=metrics_i
  fig=plt.figure(figsize=(6,8))
  ax = plt.subplot(211)
  if(indexes==[3,4,6]):
    plt.plot(columns,A.loc[A['metric']==metrics_i[0]].values[0][1:],'-o',label=metrics_i[0])
    plt.plot(columns,A.loc[A['metric']==metrics_i[1]].values[0][1:],'-o',label=metrics_i[1])
    plt.plot(columns,A.loc[A['metric']==metrics_i[2]].values[0][1:],'-o',label=metrics_i[2],color='tab:gray')
    plt.title('Metrics: '+metrics_i[0]+', '+metrics_i[1],fontsize=12)
    plt.suptitle(city,fontsize=17, y=0.94,color='tab:red',)
    plt.grid(axis='y');
    plt.yticks(np.arange(limits_y_[0],limits_y_[1],limits_y_[2]))
    ax.axes.get_xaxis().set_visible(False)
    ax.margins(0.15)
    plt.legend(fontsize=10)
    table(ax, B,rowColours=['tab:blue','tab:orange','tab:gray'], loc='bottom')
    plt.savefig('graf_'+city+'_'+metrics_i[0]+'_'+metrics_i[1]+'_'+metrics_i[2]+'.png',dpi=300)
    plt.show()
  else:
    plt.plot(columns,A.loc[A['metric']==metrics_i[0]].values[0][1:],'-o',label=metrics_i[0])
    plt.plot(columns,A.loc[A['metric']==metrics_i[1]].values[0][1:],'-o',label=metrics_i[1])
    plt.title('Metrics: '+metrics_i[0]+', '+metrics_i[1],fontsize=12)
    plt.suptitle(city,fontsize=17, y=0.94,color='tab:red',)
    plt.grid(axis='y');
    plt.yticks(np.arange(limits_y_[0],limits_y_[1],limits_y_[2]))
    ax.axes.get_xaxis().set_visible(False)
    ax.margins(0.15)
    plt.legend(fontsize=10)
    table(ax, B,rowColours=['tab:blue','tab:orange'], loc='bottom')
    plt.savefig('graf_'+city+'_'+metrics_i[0]+'_'+metrics_i[1]+'.png',dpi=300)
    plt.show()
```
