```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def Euler(F,X_0,h,I):
  #I es el intervalo donde se quiere resolver. I=(a,b)
  t=np.arange(I[0],I[1],h)
  sol=[X_0]
  #print(t)
  for t_i in t[0:-1]:
    last=sol[-1]
    sol.append(last+F(t_i,last)*h)
  return sol
```

```python
def F(t,x):
  return -x+2*np.cos(t)
X_0=1
h=0.1
I=[0,10]
t=np.arange(I[0],I[1],h)
plt.plot(t,Euler(F,X_0,h,I))
x=np.linspace(0,10,100)
plt.plot(x,np.sin(x)+np.cos(x))
```
