```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x=np.linspace(-25,220,1000)
plt.plot(x,-x/3+50,label=r'$y=-x/3+50$')
plt.plot(x,-2*x+100,label=r'$y=-2x+100$')
plt.plot([0]*2,[-2*x[-1]+100,x[-1]],label=r'$x=0$')
plt.plot(x,[0]*len(x),label=r'$y=0$')
plt.annotate(r'$R_1$',
             xy=(-10, 120),
             xytext=(1,1),
             textcoords='offset points')
plt.annotate(r'$R_2$',
             xy=(50, 100),
             xytext=(10,10),
             textcoords='offset points')
plt.annotate(r'$R_3$',
             xy=(-10, 70),
             xytext=(1,1),
             textcoords='offset points')
plt.annotate(r'$R_4$',
             xy=(0, 50),
             xytext=(10,1),
             textcoords='offset points')
plt.annotate(r'$R_5$',
             xy=(-10, 0),
             xytext=(1,5),
             textcoords='offset points')
plt.annotate(r'$R_6$',
             xy=(0/2, 0/2),
             xytext=(10,5),
             textcoords='offset points')
plt.annotate(r'$R_7$',
             xy=(45,0),
             xytext=(10,5),
             textcoords='offset points')
plt.annotate(r'$R_8$',
             xy=(-20,-70),
             xytext=(10,10),
             textcoords='offset points')
plt.annotate(r'$R_9$',
             xy=(20,-70),
             xytext=(10,10),
             textcoords='offset points')
plt.annotate(r'$R_{10}$',
             xy=(140, -70),
             xytext=(10,10),
             textcoords='offset points')
plt.annotate(r'$R_{11}$',
             xy=(215, -20),
             xytext=(0,0),
             textcoords='offset points')
x,y = np.meshgrid(np.linspace(-10,200,10),np.linspace(-150,150,10))

u = x*(-x-3*y+150)
v = y*(-2*x-y+100)


plt.quiver(x,y,u,v)
plt.axis(xmin=-20,xmax=200,ymin=-150,ymax=150)
```

```python
x=np.linspace(-25,100,1000)
plt.plot(x,10-x,label=r'$y=10-x$')
plt.plot(x,30-2*x,label=r'$y=30-2x$')
plt.plot([0]*2,[-2*x[-1]+100,x[-1]],label=r'$x=0$')
plt.plot(x,[0]*len(x),label=r'$y=0$')

x,y = np.meshgrid(np.linspace(-10,40,10),np.linspace(-40,45,10))

u = x*(10-x-y)
v = y*(30-2*x-y)


plt.quiver(x,y,u,v)
plt.axis(xmin=-10,xmax=40,ymin=-50,ymax=50)
```

```python
from scipy.integrate import solve_ivp

plt.style.use('seaborn-poster')

%matplotlib inline

F = lambda t, s: np.dot(np.array([[0, t**2], [-t, 0]]), s)
t_eval = np.arange(0, 200, 0.1)
sol = solve_ivp(F, [0, np.pi], [0], t_eval=t_eval)
plt.figure(figsize = (12, 4))
plt.subplot(121)
plt.plot(sol.t, sol.y[0])
plt.xlabel('t')
plt.ylabel('S(t)')
plt.subplot(122)
plt.plot(sol.t, sol.y[0] - np.sin(sol.t))
plt.xlabel('t')
plt.ylabel('S(t) - sin(t)')
plt.tight_layout()
plt.show()
```
