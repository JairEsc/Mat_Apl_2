```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x=np.linspace(-25,220,1000)
plt.plot(x,-x/3+50,label=r'$y=-x/3+50$')
plt.plot(x,-2*x+100,label=r'$y=-2x+100$')
plt.plot([0]*2,[-2*x[-1]+100,x[-1]],label=r'$x=0$')
plt.plot(x,[0]*len(x),label=r'$y=0$')
plt.annotate(r'$R_1$',
             xy=(-10, 120),
             xytext=(1,1),
             textcoords='offset points')
plt.annotate(r'$R_2$',
             xy=(50, 100),
             xytext=(10,10),
             textcoords='offset points')
plt.annotate(r'$R_3$',
             xy=(-10, 70),
             xytext=(1,1),
             textcoords='offset points')
plt.annotate(r'$R_4$',
             xy=(0, 50),
             xytext=(10,1),
             textcoords='offset points')
plt.annotate(r'$R_5$',
             xy=(-10, 0),
             xytext=(1,5),
             textcoords='offset points')
plt.annotate(r'$R_6$',
             xy=(0/2, 0/2),
             xytext=(10,5),
             textcoords='offset points')
plt.annotate(r'$R_7$',
             xy=(45,0),
             xytext=(10,5),
             textcoords='offset points')
plt.annotate(r'$R_8$',
             xy=(-20,-70),
             xytext=(10,10),
             textcoords='offset points')
plt.annotate(r'$R_9$',
             xy=(20,-70),
             xytext=(10,10),
             textcoords='offset points')
plt.annotate(r'$R_{10}$',
             xy=(140, -70),
             xytext=(10,10),
             textcoords='offset points')
plt.annotate(r'$R_{11}$',
             xy=(215, -20),
             xytext=(0,0),
             textcoords='offset points')
x,y = np.meshgrid(np.linspace(-10,200,10),np.linspace(-150,150,10))

u = x*(-x-3*y+150)
v = y*(-2*x-y+100)


plt.quiver(x,y,u,v)
plt.axis(xmin=-20,xmax=200,ymin=-150,ymax=150)
```

```python
x=np.linspace(-25,100,1000)
plt.plot(x,10-x,label=r'$y=10-x$')
plt.plot(x,30-2*x,label=r'$y=30-2x$')
plt.plot([0]*2,[-2*x[-1]+100,x[-1]],label=r'$x=0$')
plt.plot(x,[0]*len(x),label=r'$y=0$')

x,y = np.meshgrid(np.linspace(-10,40,10),np.linspace(-40,45,10))

u = x*(10-x-y)
v = y*(30-2*x-y)


plt.quiver(x,y,u,v)
plt.axis(xmin=-10,xmax=40,ymin=-50,ymax=50)
```

```python
fig = plt.figure()
ax = plt.axes(projection='3d')

plt.rcParams['legend.fontsize'] = 10


# First constraint
g2 = np.linspace(0.1,50,20)
g3 = np.linspace(0.1,50,20)
G2,G3 = np.meshgrid(g2,g3)
G4_1 = G2/G3/1000
ax = fig.gca(projection='3d')
c1 = ax.plot_surface(G2, G3, G4_1, label = "c1")
c1._facecolors2d=c1._facecolors3d
c1._edgecolors2d=c1._edgecolors3d

# Second
G3, G4 = np.meshgrid(g2, g3)
G2 = G3
c2 = ax.plot_surface(G2, 0*G3, G4, label = "c2")
c2._facecolors2d=c2._facecolors3d
c2._edgecolors2d=c2._edgecolors3d

# Third
G2,G3 = np.meshgrid(g2,g3)
c3 = ax.plot_surface(0*G2, G3, G2, label = "c3")
c3._facecolors2d=c3._facecolors3d
c3._edgecolors2d=c3._edgecolors3d

# And forth
#G4 = (1.04903810567666 - (0.288675134594813*G2 + 0.288675134594813*G3))/0.577350269189626
c4 = ax.plot_surface(G2, G3, G4*0, label="c4")

c4._facecolors2d=c4._facecolors3d
c4._edgecolors2d=c4._edgecolors3d

ax.legend() # -> error : 'AttributeError: 'Poly3DCollection' object has no attribute '_edgecolors2d''


# labeling the figure
#plt.xlabel('g2', fontsize=14)
#plt.ylabel('g3', fontsize=14)
ax.set_xlabel(r'$g_2$', fontsize=15, rotation=60)
ax.set_ylabel('$g_3$', fontsize=15, rotation=60)
ax.set_zlabel('$g_4$', fontsize=15, rotation=60)

plt.savefig('Constraints.jpg')
plt.show()
```
