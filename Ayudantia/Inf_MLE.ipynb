```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.special
```

```python
N=200
t=20
k=30
np.random.seed(0)
plt.figure(figsize=(10, 10))
ubicaciones=[np.random.uniform(0,1,size=t,),np.random.uniform(0,1,size=t,)]
plt.scatter(ubicaciones[0],ubicaciones[1],marker='$\U0001F63A$',c='orange',s=2**10)
```

```python
np.random.seed(10)
r=np.random.hypergeometric(t,N-t,k)
def L(N):
    return scipy.special.binom(t,r)*scipy.special.binom(N-t,k-r)/(scipy.special.binom(N,k))
plt.plot(np.arange(t+k-r,200),L(np.arange(t+k-r,200)),'.b',label=r'$L(N)$')
N_estim=np.round(t*k/r)
plt.plot([N_estim,N_estim],[0,L(N_estim),],'-g',label='argmax L(N)')
plt.legend()
print("Estimacion de N: ",N_estim)
```

```python
estimaciones=[]
for w in range(100):
    r=np.random.hypergeometric(t,N-t,k)
    if(r==0):
        continue
    else:
        estimaciones.append(np.round(t*k/r))
print(np.mean(estimaciones))
```
