```python
def dictify(book_name):
  text = book_name
  words = text.split()
  lengths = {}
  frecuencies={}
  for word in words:
      length = len(word)
      if length not in lengths:
          lengths[length] = 1
      else:
          lengths[length] += 1
  sd = sorted(lengths.items())
  return sd
```
