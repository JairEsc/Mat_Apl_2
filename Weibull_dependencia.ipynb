Dependencia de los tiempos desde última reacción para el tiempo del siguiente evento de un proceso de renovación de tiempos Weibull

```python
import numpy as np
import matplotlib.pyplot as plt
l=1
a=1.5
def psi(s,t):
  return(l*a*(l*(s+t))**(a-1)*np.exp((l*t)**(a)-(l*(s+t))**(a)))
```

```python
S=np.linspace(0,2,100)
T=[0.0001,1,2,10]
for t in T:
  plt.plot(S,psi(S,t=t))
plt.show()
```
