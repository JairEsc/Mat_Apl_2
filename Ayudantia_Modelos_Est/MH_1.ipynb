```python
#Monty Hall.
import numpy as np
import matplotlib.pyplot as plt

np.random.seed(0)#semilla
def estrategia(num_sim):
    count=0 #inicamos contador
    i=0
    proporciones_n=[]
    while(i<num_sim):
        puertas=[1,2,3]#Establecemos puertas
        puerta_premio=np.random.randint(1,len(puertas)+1) #Puerta con premio
        puerta_eleccion=np.random.randint(1,1+len(puertas)) # Elegimos puerta
        #establecemos puerta a revelar
        puertas.remove(puerta_premio)
        if(puerta_eleccion in puertas):
            count+=1#En este caso gana
        #else:#pierde
        i+=1
        proporciones_n.append((count/i))
    return proporciones_n # prop. de juegos ganados cambiando puerta
plt.plot(estrategia(10000))
plt.xlabel('Simulaciones')
plt.ylabel('Proporcion de victorias')
plt.plot([0,10000],[0.666,0.666],'--g')
plt.show()
```
