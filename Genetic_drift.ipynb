```python
import numpy as np
import matplotlib.pyplot as plt
N=1000#Poblacion
G=1500#Numero de generaciones
np.random.seed(1)
for k in range(5):
  i_0=450#Frecuencia AA inicial.
  Trayectoria=[i_0]
  for i in range(G):
      if(i_0==0 or i_0==N):
          break
      else:
          i_0=np.random.binomial(N,(i_0)/N,size=1)[0]
          Trayectoria.append(i_0)
  plt.plot(Trayectoria)
plt.show()
```

```python
#Con mutacion
alpha=0.1
beta=0.001
i_0=300
N=1000
np.random.seed(0)
Trayectoria=[i_0]
for i in range(10**3):
    p_i=(1-alpha)*(i_0/N)+beta*(1-i_0/N)
    i_0=np.random.binomial(N,p_i,size=1)[0]
    Trayectoria.append(i_0)
plt.plot(Trayectoria,'black',alpha=0.25)
plt.plot([0,10**3],[N*(beta/(alpha+beta)),N*(beta/(alpha+beta))],'g-',label=r'$\frac{2N\beta}{\alpha+\beta}$')
plt.legend()
plt.show()
```
