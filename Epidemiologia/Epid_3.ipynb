```python
import numpy as np
import matplotlib.pyplot as plt
```

Definimos una función que dado un sistema dinámico (F), una condición inicial $(X_0)$, un tamaño de paso ($h$) y un intervalo de tiempo ($I=[a,b]$), utiliza el método de Euler para aproximar la solución ($x(t)$)

```python
def Euler(F,X_0,h,I):
  #I es el intervalo donde se quiere resolver. I=(a,b)
  t=np.arange(I[0],I[1],h)
  sol=[X_0]
  #print(t)
  for t_i in t[0:-1]:
    last=sol[-1]
    sol.append(last+F(t_i,last)*h)
  return sol
```

Definimos la función de nuestro ejemplo. $X'(t)=-X(t)+2cos(t), $ $X(0)=1$.

```python
def F(t,x):
  return -x+2*np.cos(t)
```

Damos valor al tamaño de paso y definimos el intervalo de interés.

```python
X_0=1
for h in [0.05,0.02,0.01]:
    I=[0,10]
    t=np.arange(I[0],I[1],h)
    aproximacion=Euler(F,X_0,h,I)
    x=np.arange(0,10,h)
    plt.plot(x,(np.sin(x)+np.cos(x))-aproximacion,label='Error con solución real')
    plt.plot(x,np.array([round(np.sin(x)+np.cos(x),8) for x in x])-aproximacion,label='Error con solución truncada a 8 digitos')
    plt.plot(x,np.array([round(np.sin(x)+np.cos(x),5) for x in x])-aproximacion,label='Error con solución truncada a 5 digitos')
    plt.title(r'Con $h=$'+str(h))
    plt.legend()
    plt.show()
```
